```python
soup = BeautifulSoup(open("gen1.html", 'rb').read())
print(soup)
```

```python
dex = {}
num_skip = 2
for a in soup.find_all('a'):
    link = a.attrs['href'][1:].split('/')
    if 'pokemon' in link and len(link) == 2 and 'shtml' not in link[-1]:
        if num_skip == 0:
            dex[len(dex) + 1] = link[-1].upper()
            print('added to dex', link[-1])
            num_skip += 1
        else:
            num_skip -= 1
```

```python
json.dump(dex, open('../data/dex.json', 'w'))
```
