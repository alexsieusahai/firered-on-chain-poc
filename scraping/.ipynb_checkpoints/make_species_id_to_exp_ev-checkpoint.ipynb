```python
soup = BeautifulSoup(open("exp_ev.html", 'rb').read())
print(soup)
```

```python
exp = {}
ev = {}

table = soup.find_all('table')[0]
for tr in table.find_all('tr')[1:5]:
    tds = tr.find_all('td')
    try:
        speciesId = int(tds[0].text)
    except ValueError:
        continue
    print(speciesId)
    exp[speciesId] = int(tds[3].text)
    ev[speciesId] = tuple(int(tds[i].text) for i in range(4, 10))
#     print(ev[speciesId])
#     print(exp[speciesId])
    print('ingested species', speciesId)
```

```python
json.dump(exp, open('../data/exp.json', 'w'))
json.dump(ev, open('../data/ev.json', 'w'))
```
