```python
soup = BeautifulSoup(open("moves.html", 'rb').read())
print(soup)
```

```python
moves = []

table = soup.find_all('table')[0]
table = soup.find_all('table')[0]
for tr in table.find_all('tr')[2:-1]:
    tds = tr.find_all('td')
    if tds[-1].text == 'IV\n':
        break
    move = list(tds[i].text for i in range(1, 7))
    print(move)
    move[0] = move[0][:-1]
    move[1] = move[1][:-1]
    move[2] = True if move[2] == 'Physical\n' else False
    move[3] = int(move[3][:move[3].find('*')])
    try:
        move[4] = int(move[4][:move[4].find('*')])
    except ValueError:
        print(move[4])
        if move[4] == '—\n' or move[4] == '—*\n':
            move[4] = 0
        else:
            assert False
    try:
        move[5] = int(move[5][:move[5].find('%')])
    except ValueError:
        print(move[5])
        if move[5] == '—\n' or move[5] == '—*\n':
            move[5] = 100
        else:
            assert False
    moves.append(move)
    print(moves[-1])
```

```python
json.dump(moves, open('../data/moves.json', 'w'))
```
